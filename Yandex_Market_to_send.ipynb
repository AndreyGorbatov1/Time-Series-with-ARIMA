Для получения прогноза аудитории на декабрь 2020 года будем использовать прогнозирование временных рядов.

Для дальнейшего анализа воспользуемся декомпозицие временных рядов, чтобы понять характеристики изучаемого ряда.

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 8

decomposition = sm.tsa.seasonal_decompose(df['Daily_users'], model='additive')
fig = decomposition.plot()
plt.show()
```

Функция смогла уловить тренд, но не уловила сезонности, хотя о ней можно говорить, судя по графику. Хорошая модель должна была бы предсказать как продолжение тренда, так и появляение сезонности в период ноябрь-февраль.

Будем рассматривать модель ARIMA, подберём параметры, которые лучше всего будут подходить для прогнозирования нашего ряда.

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(df['Daily_users'],
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

ARIMA(0, 1, 1)x(0, 1, 0, 12) - наилучший набор параметров. Натренируем модель на наших данных с применение оптимального набора параметров.

```python
mod = sm.tsa.statespace.SARIMAX(df['Daily_users'],
                                order=(0, 1, 1),
                                seasonal_order=(0, 1, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()

print(results.summary().tables[1])
```

```python
results.plot_diagnostics(lags=2,figsize=(16, 8))
plt.show()
```

Посмотрим, насколько точно модель предсказывает на исторических данных:

```python
pred = results.get_prediction(start=pd.to_datetime('2019-01-01'), end=pd.to_datetime('2020-01-01') , dynamic=False)
pred_ci = pred.conf_int()

ax = df['Daily_users'].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Daily_users')
plt.legend()
plt.show()
```

Кроме пары выбросов, модель достаточно точно предсказала аудиторию на исторических данных.

```python
y_forecasted = pred.predicted_mean
y_truth = df['Daily_users']['2019-01-01':'2020-01-01']
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```

Используем модель для предсказания аудитории в будущем. 

```python
pred_uc = results.get_forecast(steps = 15)

pred_ci = pred_uc.conf_int()

ax = df['Daily_users'].plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Daily_users')

plt.legend()
plt.show()
```

```python
print('Number of predicted daily users on 2020-12-01: %.0f ±%.0f' % (pred_uc.predicted_mean['2020-12-01'], pred_uc.predicted_mean['2020-12-01']-pred_ci['lower Daily_users']['2020-12-01']))
```

Как видно, полученная модель повторяет тренд исторических данных и влючает в себя ожидаемую сезонность.
